```python
import re
```

```python
pattern = re.compile("text")
```

## Two types of Programs- 1. Match, 2. Search

#### `Match()` looks for the pattern at the beginning of the string

```python
print(pattern.match("boo"))
```

```python
print(pattern.match("text"))
```

```python
print(pattern.match("text object"))
```

```python
print(pattern.match("object text"))
```

#### `Search()` scans through the string looking for all matches

```python
print(pattern.search("text object"))
```

```python
print(pattern.search("object text"))
```

```python
print(pattern.search("text object text"))
```

Parameters

For example `@[A-Za-z0-9_]+` will return all words that:

* start with `@` and are followed by at least one:
* capital letter (`A-Z`)
* lowercase letter (`a-z`)
* number (`0-9`)
* or underscore (`_`)

```python
text = '@UN @UN_Women "Ethics are built right into the ideals and objectives of the United Nations" \
#UNSG @ NY Society for Ethical Culture bit.ly/2guVelr'
split_text = text.split(' ')
```

```python
split_text
```

```python
[w for w in split_text if re.search('@[A-Za-z0-9_]+', w)]
```

```python
date_text = "03/25/93 Total time of visit (in minutes):\n"
```

```python
date_time_series = pd.Series(doc)
```

```python
date_time_series
```

```python
date_df = pd.DataFrame(date_time_series, columns=["Appointments"])
```

```python
date_df.head()
```

```python
date_df['Date'] = date_df['Appointments']
```

```python
def date_finder(x):
    string = x
    try:
        date = re.findall(r'([0-9]*/[0-9]*/[0-9]*)', x)
#         print(date)
        return date
    except :
        print("No date found")
```

```python
date_df['Date'].head()
```

```python
date_df['Parsed_Date'] = date_df['Date'].apply(date_finder)
```
